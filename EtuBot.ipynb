# FUNCTIONS FOR FINANCIAL EVALUATION

```python
def pltcolor(lst):
    cols=[]
    for i in range(lst.shape[0]):
        if lst.iloc[i] == -1:
            cols.append('green')
        elif lst.iloc[i] == 1:
            cols.append('red')
    return cols


def financial_evaluation(prices,predictions,is_there_short,capital,commission):
    """
    prices: Adjusted close price data.
    predictions: Signals generated by our strategy. -1: Buy, 0: Hold, 1: Sell
    is_there_short: True if we open the short positions when a sell signal comes, False if we don't.
    capital: Initial capital amount when we started to trade. We assume that we use all the money we have for each trade.
    commission: Amount of dollar we paid to brokers for opening each trade.    
    """
    print("------------------")
    initial_capital = capital
    print(f"Inital capital Amount = {initial_capital}, commission = {commission}$ and short {'is used' if is_there_short else 'is not used'}")
    print("-------------")
    long = False
    short = False
    profit_trade = 0
    loss_trade = 0
    longPrice = 0.0
    shortPrice = 0.0
    total_day_position_open = 0
    max_profit = 0.0
    max_loss = 0
    max_capital = initial_capital
    min_capital = initial_capital
    RoR = 0.0  
    if is_there_short == False:
        for i in range(len(predictions)):
            change = 0
            if predictions[i] == -1 and long == False:
                long = True
                longPrice = prices.iloc[i]  
                capital -= commission
            elif predictions[i] == 1 and long == True:
                long = False
                change = (prices.iloc[i] - longPrice) / longPrice * 100
                if change > 0:
                    profit_trade += 1
                elif change < 0:
                    loss_trade += 1
                capital *= 1 + change*0.01
                capital -= commission
                if change > max_profit:
                    max_profit = change
                if change < max_loss:
                    max_loss = change
                if capital > max_capital:
                    max_capital = capital
                if capital < min_capital:
                    min_capital = capital
            if long == True:
                total_day_position_open += 1
    elif is_there_short == True:
        for i in range(len(predictions)):
            change = 0
            if predictions[i] == -1 and long == False:
                if short == True:
                    change = (shortPrice-prices.iloc[i]) / shortPrice * 100
                    if change > 0:
                        profit_trade += 1
                    elif change < 0:
                        loss_trade += 1      
                    capital *= 1 + change*0.01  
                    capital -= commission
                long = True
                short = False
                longPrice = prices.iloc[i]
                capital -= commission
            elif predictions[i] == 1 and short == False:
                if long == True:
                    change = (prices.iloc[i]-longPrice) / longPrice * 100
                    if change > 0:
                        profit_trade += 1
                    elif change < 0:
                        loss_trade += 1
                    capital *= 1 + change*0.01  
                    capital -= commission
                short = True
                long = False
                shortPrice = prices.iloc[i]
                capital -= commission
            if change > max_profit:
                max_profit = change
            if change < max_loss:
                max_loss = change
            if capital > max_capital:
                max_capital = capital
            if capital < min_capital:
                min_capital = capital
            if long == False and short == False:
                total_day_position_open += 1
    total_day_position_open = len(prices) - total_day_position_open
    print(f"Buy&Hold return: %{ (prices[-1] - prices[0]) / prices[0] * 100} ")
    RoR = (capital - initial_capital) / initial_capital * 100
    print(f"Total return: %{RoR} ")
    year_count = len(prices) / 365
    print(f"Annualized return: %{RoR/year_count}")
    print(f"Average return per trade: %{RoR/(profit_trade + loss_trade)}.")
    print(f"Total trade count: {profit_trade + loss_trade}.")
    print(f"Profitable trade count: {profit_trade} and loss trade count: {loss_trade}.")
    print(f"Annualized total trade count: {(profit_trade + loss_trade) / year_count}.")
    print(f"Average trade day: {total_day_position_open/(profit_trade + loss_trade)}.")
    print(f"Max profit for a trade: %{max_profit}.")
    print(f"Max loss for a trade: %{max_loss}.")
    print(f"Max capital value: {max_capital}.")
    print(f"Min capital value: {min_capital}.")
    
    plt.figure(figsize=(20, 10), dpi=80)
    prices.plot(kind="line", stacked=False,linewidth=1)
    plt.scatter(x = predictions[predictions!=0].index ,y = prices[predictions!=0],s=20,c=pltcolor(predictions[predictions!=0]))
    plt.grid(True)
    plt.show()    
```

# STRATEGY #1

```python
adjClose = adjClose.assign(Label=np.zeros((len(adjClose),1)))

adjClose["200 MA"] = adjClose["Adj Close"].rolling(200).mean()
adjClose["50 MA"] = adjClose["Adj Close"].rolling(50).mean()
adjClose["20 MA"] = adjClose["Adj Close"].rolling(20).mean()

for i in range(len(adjClose)):
    if adjClose.at[adjClose.index[i],"50 MA"] > adjClose.at[adjClose.index[i],"200 MA"] and adjClose.at[adjClose.index[i],"20 MA"] > adjClose.at[adjClose.index[i],"50 MA"]:
        adjClose.at[adjClose.index[i],"Label"] = -1
    if adjClose.at[adjClose.index[i],"50 MA"] < adjClose.at[adjClose.index[i],"200 MA"] and adjClose.at[adjClose.index[i],"20 MA"] < adjClose.at[adjClose.index[i],"50 MA"]:
        adjClose.at[adjClose.index[i],"Label"] = 1

financial_evaluation(adjClose["Adj Close"],adjClose["Label"],False,10_000,1)
financial_evaluation(adjClose["Adj Close"],adjClose["Label"],True,10_000,1)
```
